```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 

from pandas.plotting import scatter_matrix 
```

```python
# 8.(a) Read college data set 

df = pd.read_csv("College.csv")
```

```python
# 8.(b) 

df.head() # print first 5 rows in a dataframe
```

```python
#df.rename(columns={"Unnamed: 0" : "College_Name"}, inplace=True) --> it will also work 

df.rename(columns={"None" : "College_Name"}, inplace=True) # rename the column

df.head()
```

```python
# 8. (c) 

df.describe() # the summary of the variables
```

```python
# find the scatterplot matrix of the first ten columns of the variables of the data 

scatter_matrix(df.iloc[:, :10], figsize=(12,12), diagonal="hist", alpha=0.5)

plt.suptitle("scatterplot matrix of first 10 columns")
plt.show()
```

```python
# side-by-side boxplots of Outstate versus Private

sns.boxplot(data=df, x="Outstate", y="Private")

plt.suptitle("Boxplots of Outstate versus Private")
plt.show()
```

```python
# create a new qualitative variable, called Elite, by binning the Top10perc variable
# we are going to divide universities into two groups based on whether or not the Top10perc exceeds 50 %

df["Elite"] = df["Top10perc"] > 50

df.head()
```

```python
# find the number of elite universities 

true_count = df["Elite"].sum()

print(true_count)
```

```python
# side-by-side boxplots of Outstate versus Elite

sns.boxplot(data=df, y="Outstate", x="Elite")

plt.suptitle("Boxplots of Outstate versus Elite")
plt.show()
```

```python
# produce some histograms with differing numbers of bins 

df["Books"].hist(bins=5)
plt.title("Histogram with 5 bins")
plt.show()
```

```python
# 9.(a) 

df = pd.read_csv("Auto.csv")

df.head()
```

```python
print(df.dtypes)

# numeric types (int64, float64) are usually quantitative

# object or category types are usually qualitative
```

```python
# find the range of each quantitative predictor 

quantitative_cols = df.select_dtypes(include=['number'])

ranges = quantitative_cols.agg(['min', 'max']) 
ranges.loc['range'] = ranges.loc['max'] - ranges.loc['min'] # adds new row

print(ranges)

ranges = quantitative_cols.agg(['min', 'max']).T  # transpose for easier reading
ranges['range'] = ranges['max'] - ranges['min'] # adds new column

print(ranges)
```

```python
# mean and standard deviation of each quantitaive predictor

stats = quantitative_cols.agg(['mean', 'std']).T

print(stats)
```

```python
new_df = df.drop(df.index[10:86])

quantitative_cols = new_df.select_dtypes(include="number")

stats = quantitative_cols.agg(['min', 'max', 'mean', 'std']).T

stats['range'] = stats['max'] - stats['min']

stats.drop(columns='min', inplace=True) # inplace is to update original dataframe 

#stats = stats.drop(columns='min') --> also works

stats.drop(columns='max', inplace=True)

print(stats)
```
